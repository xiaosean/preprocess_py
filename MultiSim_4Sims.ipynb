```python
import pandas as pd
import numpy as np
```

```python
def check_ci_bound(data, ci=0.95):
    _data = pd.DataFrame(np.sort(data).tolist(), columns = ["value"])
    _data['interval'] = np.ceil(_data['value']/0.05)*0.05
    return _data.iloc[int(np.floor(ci * len(_data) )), 1]
```

```python
cols = ['DATA_MONTH', 'MINING_DW_SUBSCR_NO', 'DW_PARTY_ID', 'PTY_CBU_PO_CNT', 'SUBSCR_STATUS_CODE', 'DORMANT_MONTH_CNT_MKT', 'TENURE_SCV', 'RPS_NAME', 'CUST_TYPE']
multisim_mds = pd.read_csv('NEW_DATA_FULL_2017_10_18/ALL_MDS.txt', usecols=cols)
```

```python
multisim_mds_filter = multisim_mds[(multisim_mds.RPS_NAME == "CONSUMER MOBILITY") & (multisim_mds.SUBSCR_STATUS_CODE == 'A') & \
            (multisim_mds.TENURE_SCV >=3) & (multisim_mds.DORMANT_MONTH_CNT_MKT < 2) & \
            (multisim_mds.DATA_MONTH == '2017/4/1')]
```

# Location

```python
df4 = pd.read_csv('NEW_DATA_FULL_2017_10_18/OH_FNL_4_4SIM_20171013.txt', error_bad_lines=False)
```

```python
sims = ['A', 'B', 'C', 'D']
for s in sims:
    df4[s + '_LOCATION'] = df4[s + '_LOCATION_DATA'] + df4[s + '_LOCATION_VOICE']
```

```python
sims = ['A', 'B', 'C', 'D']
for i in range(len(sims) - 1):
    s1 = sims[i]
    for s2 in sims[i+1:]:
        df4[s1 + s2 + '_INTERSECTION'] = df4[s1 + s2 + '_INTERSECTION_VOICE'] + df4[s1 + s2 + '_INTERSECTION_DATA']
        df4[s1 + s2 + '_intersect_persentage'] = df4[s1 + s2 + '_INTERSECTION'] / ((df4[s1 + '_LOCATION'] + df4[s2 + '_LOCATION']) / 2)

df4 = df4.fillna(0)
```

```python
df4[df4.columns[-6:]].describe()
```

```python
location_intersection = pd.concat([df4.loc[:, 'AB_intersect_persentage'], 
                               df4.loc[:, 'AC_intersect_persentage'],
                               df4.loc[:, 'AD_intersect_persentage'], 
                               df4.loc[:, 'BC_intersect_persentage'], 
                               df4.loc[:, 'BD_intersect_persentage'], 
                               df4.loc[:, 'CD_intersect_persentage']])

location_bound = check_ci_bound(location_intersection)
```

```python
location_bound
```

# Same and Different

```python
df1 = pd.read_csv('NEW_DATA_FULL_2017_10_18/OH_FNL_1_4SIM_20171013.txt', error_bad_lines=False)
```

```python
df1.columns
```

```python
result = df1[['DW_PARTY_ID', 'MINING_DW_SUBSCR_NO_A', 'MINING_DW_SUBSCR_NO_B', 'MINING_DW_SUBSCR_NO_C', 'MINING_DW_SUBSCR_NO_D']]
```

```python
sims = ['A', 'B', 'C', 'D']
for i in range(len(sims) - 1):
    s1 = sims[i]
    for s2 in sims[i+1:]:
        result.loc[:, s1 + s2 + '_diff'] = ((df2[s1 + s2 + '_intersect_persentage'] < call_bound) & \
                                    (df4[s1 + s2 + '_intersect_persentage'] < location_bound)) | \
                                    ((df2[s1 + s2 + '_intersect_persentage'] < call_bound) & \
                                    (df4[s1 + s2 + '_intersect_persentage'] >= location_bound) & \
                                    (df1[s1 + s2 + '_SELF_CALL'] > 0))
```

```python
result = pd.merge(result, mining_id_df, how='left', on='DW_PARTY_ID')
```

```python
result.MINING_DW_SUBSCR_NO.isnull().sum()
```

```python
result = result.dropna()
```

```python
sims = ['A', 'B', 'C', 'D']
result['All_diff'] = True
for index, row in result.iterrows():
    flag = True
    for i in range(len(sims)-1):
        s1 = sims[i]
        if row['MINING_DW_SUBSCR_NO_' + s1] in row['MINING_DW_SUBSCR_NO']:
            for s2 in sims[i+1:]:
                if row['MINING_DW_SUBSCR_NO_' + s2] in row['MINING_DW_SUBSCR_NO'] and row[s1 + s2 + '_diff'] == False:
                    flag = False
                    break
        if flag == False:
            result.at[index, 'All_diff'] = False
            break
```

```python
result.All_diff.value_counts()
```

```python
result_diff_id_df = pd.DataFrame()
result_diff_id_df['MINING_DW_SUBSCR_NO'] = np.concatenate(result[result.All_diff].MINING_DW_SUBSCR_NO.values)

result_other_id_df = pd.DataFrame()
result_other_id_df['MINING_DW_SUBSCR_NO'] = np.concatenate(result[result.All_diff != True].MINING_DW_SUBSCR_NO.values)
```
