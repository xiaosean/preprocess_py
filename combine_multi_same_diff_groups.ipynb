```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('D:/NEW_DATA_FULL_10_6/MDS_MULTI_PTYCBU_1_4.txt', error_bad_lines=False, 
                 usecols=['MINING_DW_SUBSCR_NO', 'DW_PARTY_ID', 'TENURE_SCV', 'DORMANT_MONTH_CNT_MKT', 'SUBSCR_STATUS_CODE'])
```

```python
len(df)
```

```python
df = df[df['DORMANT_MONTH_CNT_MKT'] < 2]
df = df[df['TENURE_SCV'] >= 3]
df = df[df['SUBSCR_STATUS_CODE'] == 'A']
```

```python
len(df)
```

```python
path = "./"
filename = "1006_segment_same_multisim.csv"
# TAG K MEAN RESULT
GROUP_ID_FILE = path + filename
group_df_same = pd.read_csv(GROUP_ID_FILE, error_bad_lines=False, sep=' ', header=None)
group_df_same.columns = ['DW_PARTY_ID', 'Groups', 'sim_count']
```

```python
path = "./"
filename = "1006_segment_diff_multisim.csv"
# TAG K MEAN RESULT
GROUP_ID_FILE = path + filename
group_df_diff = pd.read_csv(GROUP_ID_FILE, error_bad_lines=False, sep=' ', header=None)
group_df_diff.columns = ['MINING_DW_SUBSCR_NO', 'Groups', 'sim_count']
```

```python
len(group_df_same)
```

```python
len(group_df_diff)
```

```python
df = pd.merge(df, group_df_diff, on='MINING_DW_SUBSCR_NO', how='left')
df
```

```python
df.Groups.isnull().sum()
```

```python
party_map = {}
for r in group_df_same.iterrows():
    party_map[r[1]['DW_PARTY_ID']] = r[1]['Groups']
```

```python
not_found_count = 0
for index, row in df.iterrows():
    try:
        group = party_map[row['DW_PARTY_ID']]
        df.set_value(index,'Groups',group)
    except KeyError:
        if type(row['Groups']) == float and np.isnan(row['Groups']):
            not_found_count += 1
print("id not found:", not_found_count)
```

```python
df = df[['MINING_DW_SUBSCR_NO', 'Groups']]
df = df.dropna()
len(df)
```
