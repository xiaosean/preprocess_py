```python
# set configure
path = "C:/Users/VIPLAB/Desktop/preprocess_py/"
filename = "0814_marketing_with_picked_group11_numeric_max_min"
out_filename = filename
relative_filename = path + filename + ".csv"
print(relative_filename)

t0 = time()
df = pd.read_csv(relative_filename, error_bad_lines=False)

print("time for read csv: %.2f" % (time()-t0))
```

```python
df_cate = df[cate_cols]
df_num = df[num_cols]
```

```python
groups = list(df_num["Groups"].unique())
all_groups = list(df_num["Groups"].unique())
for group in groups:
    dict_groups = {}
    print("\nRunning Group", group, "------------------------")
    for each_group in all_groups:
        if each_group != group:
            dict_groups[each_group] = 'Others'
        else:
            dict_groups[each_group] = group
    print("dict_groups", dict_groups)
    res_df_num = df_num.copy()
    res_df_cate = df_cate.copy()
    res_df_num["Groups"] = res_df_num["Groups"].map(dict_groups)
    res_df_cate["Groups"] = res_df_num["Groups"]
    group_count = res_df_num['Groups'].value_counts()
    print(group_count)
    
    # SMOTE num
    X, y = res_df_num.iloc[:, 0:-1], res_df_num.iloc[:, -1]
    sm = SMOTE(random_state=42, n_jobs=-1)
    print("SMOTE start")
    X_res, y_res = sm.fit_sample(X, y)
    print("SMOTE Finish")
    
    # Oversample cate
    print("cate start")
    target_group = res_df_cate[res_df_cate['Groups'] == group]
    target_total = group_count[group]
    others_total = group_count['Others']
    gen_df = pd.DataFrame()
    for c in target_group.columns[:-1]:
        vc_df = target_group[c].value_counts()
        prob = []
        for ind in vc_df.index:
            prob.append(vc_df[ind] / target_total)
        gen_col = []
        for i in range(others_total - target_total):
            rnd = random.random()
            cur_sum = 0
            for j in range(len(prob)):
                if rnd > cur_sum and rnd < cur_sum + prob[j]:
                    gen_col.append(vc_df.index[j])
                    break
                cur_sum += prob[j]
        gen_df[c] = gen_col
    gen_df['Groups'] = group
    print("cate finished")
    
    print("join df")
    res_df_cate = res_df_cate.append(gen_df)
    res_df_cate.index = list(range(len(res_df_cate)))
    res_df = pd.DataFrame(data = X_res, columns = df_num.columns[0:-1])
    for c in res_df_cate.columns[:-1]:
        res_df[c] = res_df_cate[c]
    res_df["Groups"] = y_res
    if not np.array_equal(res_df_cate['Groups'].values, res_df['Groups'].values):
        print("EEEEEEEEEERROR")
    print("output config")
    out_path = out_dir + group + "/"
    if not os.path.exists(out_path):
        os.makedirs(out_path)
    print("start write csv")
    save_dataframe(res_df, out_path + out_filename + "_sample_" + group + "_others")
```
