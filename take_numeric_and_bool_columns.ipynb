```python
interesting_list = []
for col in df.columns:
    if np.issubdtype(df[col].dtypes, np.number):
        interesting_list.append(col)
    else:
        unique_list  = df[col].unique()
        if(len(unique_list) == 2):
            for idx, item in enumerate(unique_list):
#         let true or false turn to 0 or 1
                df.loc[df[df[col] == item].index, col] = idx
            df[col] = df[col].apply(np.int64)
            interesting_list.append(col)

```
