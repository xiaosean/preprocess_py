```python
# set configure
path = "D:/NEWDATAMULTISIM_2017_9_4/"
filename = "MULTISIM_VOICEDATA"
# relative_filename = path + filename + ".csv"
relative_filename = path + filename + ".txt"
# wants_cols = ['MINING_DW_SUBSCR_NO', 'CUST_TYPE']
```

```python
# read csv file and print cost time
t0 = time()
df = pd.read_csv(relative_filename)
# df = pd.read_csv(relative_filename, error_bad_lines=False)
print("time for read csv: %.2f" % (time()-t0))
```

```python
df = df.fillna(0)
```

# fillna and aggregate

```python
party_sum_df = df[df.columns[2:]].fillna(0).groupby("DW_PARTY_ID").sum()
```

```python
party_sum_df["DW_PARTY_ID"] = party_sum_df.index

party_sum_df
```

```python
len(party_sum_df)
```

```python
value_cols = ["SUM_MOC_CNT", "SUM_MOC_DUR", "SUM_MTC_CNT", "SUM_MTC_DUR", "SUM_data_USAGE_MB"]
_normal_party_list = []
normal_party_df = party_sum_df.copy()
for col in value_cols:
    normal_party_df = normal_party_df[normal_party_df[col] > 0]
len(normal_party_df)
```

```python
normal_party_df.head()
```

```python
start_t = time()
sum_normal_party_dict = {}
for index, row in normal_party_df.iterrows():
    sum_normal_party_dict[int(row["DW_PARTY_ID"])] = {}
    for col in value_cols:
        sum_normal_party_dict[int(row["DW_PARTY_ID"])][col] = row[col]
print(time() - start_t)
```

```python
sum_normal_party_dict
```

```python
ratio_party_id = pd.merge(df, normal_party_df[['DW_PARTY_ID']], how='inner', on= 'DW_PARTY_ID')
```

```python
ratio_party_id.head()
```

```python
start_t = time()
for index, row in ratio_party_id.iterrows():
    for col in value_cols:
#         row[col] = row[col] / normal_party_df[normal_party_df['DW_PARTY_ID'] == row['DW_PARTY_ID']][col]
        row[col] = row[col] / sum_normal_party_dict[int(row['DW_PARTY_ID'])][col]
    ratio_party_id.iloc[index] = row
#         ratio_party_id.loc[ratio_party_id[ratio_party_id["DW_PARTY_ID"] == row['DW_PARTY_ID']].index, col] = row[col] / sum_normal_party_dict[int(row['DW_PARTY_ID'])][col]

print(time() - start_t)
```
