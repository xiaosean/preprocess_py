```python
avg = pd.DataFrame()

cutoff = round(len(df) * 0.05)
print("-------------%s------------" % ('all'))
avg_tmps = []
for c in df[cols].columns:
    if c == 'Groups':
        continue
    if 'DATA_USAGE' in c or 'DUR' in c:
        avg_tmp = np.mean(np.sort(df[c].values)[:-cutoff])
        avg_tmps.append("Avg: %.2f" % (avg_tmp))
        print("%s %.2f %s" % (c, avg_tmp, 'cut'))
    else:
        avg_tmp =  np.mean(df[c].values)
        avg_tmps.append("Avg: %.2f" % (avg_tmp))
        print("%s %.2f" % (c, avg_tmp))
avg['all'] = avg_tmps

for g in groups_list:
    print("-------------%s------------" % (g))
    avg_tmps = []
    for c in df[cols].columns:
        if c == 'Groups':
            continue
        if 'DATA_USAGE' in c or 'DUR' in c:
            avg_tmp = df.sort_values(by=c).iloc[:-cutoff][df['Groups'] == g][c].mean()
            avg_tmps.append("Avg: %.2f" % (avg_tmp))
            print("%s %.2f %s" % (c, avg_tmp, 'cut'))
        else:
            avg_tmp =  np.mean(df[df['Groups'] == g][c].values)
            avg_tmps.append("Avg: %.2f" % (avg_tmp))
            print("%s %.2f" % (c, avg_tmp))
    avg[g] = avg_tmps
print('writing to csv')
avg.to_csv("avgs.csv", index=False, encoding='utf-8')
print('done')
```

```python
start_time = time()
# group_name = 'all'
# path = "C:/Users/VIPLAB/Desktop/preprocess_py/marketing_analyze/0731_sample/" + group_name + "/"
# filename = '0731_marketing_with_picked_group11_numeric_max_min_sample_' + group_name + '_others_0.csv' 
# filename = '0731_marketing_with_picked_group11_numeric_max_min_sample_' + group_name + '_0.csv' 

path = 'C:/Users/VIPLAB/Desktop/preprocess_py/'
filename = '0814_marketing_with_picked_group11.csv'

df_single = pd.read_csv(path + filename, error_bad_lines=False)

print('load feather %s seconds', format(time() - start_time))
```

```python
cols = df_single.columns
cate_cols = []
for c in cols:
    if "IMEI" in c or 'ZIP' in c or 'TYPE' in c or c.endswith('ID') or "OPERATOR" in c or "NAME" in c or "DESC" in c or "IVR" in c:
#         print(c)
        cate_cols.append(c)
cate_cols.append('Groups')
cate_cols
```

```python
df_multi = df_multi[cate_cols]
df_single = df_single[cate_cols]
```

```python
df = df_multi.append(df_single)
df = df.reset_index(drop=True)
```

```python
cate_cols.remove('Groups')
```

```python
cate_cols
```

```python
df['BILL_ZIP_CODE'] = df['BILL_ZIP_CODE'].apply(pd.to_numeric, errors = 'coerce').fillna(0).astype(int).astype(str)
```

```python

total_count = len(df)
top = 3
# cate_cols = ['BILL_ZIP_CODE', 'DATA_USAGE_MB']

top3 = pd.DataFrame()
top3['column'] = cate_cols

print("-------------%s------------" % ('all'))
tops = []
for c in df[cate_cols].columns:
    if c == 'Groups':
        continue
    top_str = ""
    for i in range(top if len(df[c].value_counts()) > top else len(df[c].value_counts())):
        top_str += str(df[c].value_counts().index[i]).strip() + " "
        top_str += "%.2f%%" % (df[c].value_counts().values[i] / total_count * 100)
        if i + 1 != top:
            top_str += ", "
    print(top_str)
    tops.append(top_str)
top3['all'] = tops

for g in groups_list:
    print("-------------%s------------" % (g))
    total_count = len(df[df['Groups'] == g])
    tops = []
    for c in df[df['Groups'] == g][cate_cols].columns:
        if c == 'Groups':
            continue
        top_str = ""
        temp_vc = df[df['Groups'] == g][c].value_counts()
        for i in range(top if len(temp_vc) > top else len(temp_vc)):
            top_str += str(temp_vc.index[i]).strip() + " "
            top_str += "%.2f%%" % (temp_vc.values[i] / total_count * 100)
            if i + 1 != top:
                top_str += ", "
        print(top_str)
        tops.append(top_str)
    top3[g] = tops

    
print('writing to csv')
top3.to_csv("top3_1007.csv", index=False, encoding='utf-8')
print('done')
```
